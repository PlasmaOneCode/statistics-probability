```python
#library yang digunakan
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statistics as st
from scipy import stats
import seaborn as sns

#ekspor dataset sebagai data
data = pd.read_excel("Dataset.xlsx")
#specify kolom dan membersihkan sel kosong
dhh = pd.to_numeric(data["Curah Hujan (mm)"],errors='coerce')
daa =  pd.to_numeric(data["Penjualan A (pcs)"],errors='coerce')
dbb = pd.to_numeric(data["Penjualan B (pcs)"],errors='coerce')
dh = dhh.dropna()
da = daa.dropna()
db = dbb.dropna()

vis = {
    'Curah Hujan' : dh,
    'Penjualan A' : da,
    'Penjualan B' : db
}

print(vis)
```

```python
#mencari dan mencetak rata rata

rhujan = np.mean(dh)
rA = np.mean(da)
rB = np.mean(db)

print(f"rata-rata curah hujan: {rhujan:.2f}")
print(f"rata-rata penjualan A: {rA:.2f}")
print(f"rata-rata penjualan B: {rB:.2f}")
```

```python
#mencari dan mencetak median

medh = np.median(dh)
meda = np.median(da)
medb = np.median(db)

print(f"median curah hujan: {medh:.2f}")
print(f"median penjualan A: {meda:.2f}")
print(f"median penjualan B: {medb:.2f}")
```

```python
#mencari dan mencetak modus

modh = st.mode(dh)
moda = st.mode(da)
modb = st.mode(db)

print(f"modus curah hujan: {modh:.2f}")
print(f"modus penjualan A: {moda:.2f}")
print(f"modus penjualan B: {modb:.2f}")
```

```python
#mencari dan mencetak range

rangeh = dh.max() - dh.min()
rangea = da.max() - da.min()
rangeb = db.max() - db.min()

print(f"range curah hujan: {rangeh:.2f}")
print(f"range penjualan A: {rangea:.2f}")
print(f"range penjualan B: {rangeb:.2f}")
```

```python
#mencari dan mencetak quartil

IQRh = np.percentile(dh,75) - np.percentile(dh,25)
IQRa = np.percentile(da,75) - np.percentile(da,25)
IQRb = np.percentile(db,75) - np.percentile(db,25)
#bisa juga dengan sintaks
#IQRh = dh.quantile(0.75) - dh.quantile(0.25)
#IQRa = da.quantile(0.75) - da.quantile(0.25)
#IQRb = db.quantile(0.75) - db.quantile(0.25)

print(f"IQR curah hujan: {IQRh:.2f}")
print(f"IQR penjualan A: {IQRa:.2f}")
print(f"IQR penjualan B: {IQRb:.2f}")
```

```python
#mencari dan mencetak variansi

varh = dh.var()
vara = da.var()
varb = db.var()

print(f"varian curah hujan: {varh:.2f}")
print(f"varian penjualan A: {vara:.2f}")
print(f"varian penjualan B: {varb:.2f}")
```

```python
#mencari dan mencetak standar deviasi

stddevh = dh.std()
stddeva = da.std()
stddevb = db.std()

print(f"Standar Deviasi curah hujan: {stddevh:.2f}")
print(f"Standar Deviasi penjualan A: {stddeva:.2f}")
print(f"Standar Deviasi penjualan B: {stddevb:.2f}")
```

```python
#mencari dan mencetak skewness

skwh = stats.skew(dh)
skwa = stats.skew(da)
skwb = stats.skew(db)

print(f"Skewness curah hujan: {skwh:.2f}")
print(f"Skewness penjualan A: {skwa:.2f}")
print(f"Skewness penjualan B: {skwb:.2f}")

sns.kdeplot([dh, da, db])
```

```python
#mencari dan mencetak kurtosis

kurth = dh.kurtosis()
kurta = da.kurtosis()
kurtb = db.kurtosis()

print(f"Kurtosis curah hujan: {kurth:.2f}")
print(f"Kurtosis penjualan A: {kurta:.2f}")
print(f"Kurtosis penjualan B: {kurtb:.2f}")

sns.histplot(dh, bins=30, kde=True)
plt.title('Distribusi Curah Hujan (mm)')
plt.xlabel('Penjualan A (pcs)')
plt.ylabel('Frekuensi')
plt.show()

sns.histplot(da, bins=30, kde=True)
plt.title('Distribusi Penjualan A (pcs)')
plt.xlabel('Penjualan A (pcs)')
plt.ylabel('Frekuensi')
plt.show()

sns.histplot(db, bins=30, kde=True)
plt.title('Distribusi Penjualan B (pcs)')
plt.xlabel('Penjualan A (pcs)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
#membuat diagram frekuensi curah hujan

bins = 'auto'
freq, bin_edges = np.histogram(dh, bins=bins)

tabelF = pd.DataFrame({
    'Interval': pd.IntervalIndex.from_breaks(bin_edges),
    'Frekuensi': freq  #menggunakan frekuensi dari histogram
})

tabelF = tabelF.groupby('Interval').sum().reset_index()  #mengelompokkan berdasarkan interval

print("Tabel Distribusi Frekuensi:")
print(tabelF)

plt.figure(figsize=(8, 6))
plt.hist(dh, bins=bins, edgecolor='black', alpha=0.7)
plt.title('Histogram Curah Hujan')
plt.xlabel('Curah Hujan (mm)')
plt.ylabel('Frekuensi')
plt.grid()
plt.show()
```

```python
#membuat scatter plot dan regresi linear nya antara curah hujan dengan penjualan A

plt.figure(figsize=(8, 4))
sns.regplot(y=dh, x=da, scatter_kws={'alpha':.6}, line_kws={'color':'red'})
plt.title('Korelasi Curah Hujan dengan Penjualan A')
plt.xlabel('Penjualan A (pcs)')
plt.ylabel('Curah Hujan (mm)')
plt.grid()
plt.show()
#membuat scatter plot dan regresi linear nya antara curah hujan dengan penjualan A

plt.figure(figsize=(8, 4))
sns.regplot(y=dh, x=db, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title('Korelasi Curah Hujan dengan Penjualan B')
plt.xlabel('Penjualan B (pcs)')
plt.ylabel('Curah Hujan (mm)')
plt.grid()
plt.show()
```

```python
#membuat boxplot tiap kolom

plt.figure(figsize=(8,6))
plt.boxplot(dh, vert=False)
plt.title('Boxplot Curah Hujan')
plt.xlabel('Curah Hujan (mm)')
plt.show()

plt.figure(figsize=(8,6))
plt.boxplot(da, vert=False)
plt.title('Boxplot Penjualan A')
plt.xlabel('Penjualan A (pcs)')
plt.show()

plt.figure(figsize=(8,6))
plt.boxplot(db, vert=False)
plt.title('Boxplot Penjualan B')
plt.xlabel('Penjualan B (pcs)')
plt.show()
```

> Penjelasan mengenai boxplot diatas

Dari data tersebut, kegunaan boxplot adalah menampilkan pemusatan datanya secara ringkas yang mencangkup nilai minimal dan maksimal data, lalu median, kuartil dan outlier nya. 
Jika dilihat dari ketiga boxplot di atas, data yang diambil dari 3 sumber tersebut tentunya memiliki distribusi yang berbeda-beda.

Pada boxplot curah hujan, terdapat outlier yang berada di luar boxplot yang menandakan bahwa terdapat nilai yang berbeda atau menonjol dari nilai lainnya. Dimana data tersebut menunjukkan bahwa terdapat hari dengan curah hujan yang sangat tinggi dibandingkan dengan hari lainnya. Dari Outlier yang muncul, ada nilai ekstrim yaitu pada outlier dengan nilai kurang lebih 170an. 
//Ekstrim bagian atas apabila nilainya berada di atas Q3 + (3 x IQR)//

Kemudian pada boxplot penjualan A dan penjualan B memiliki skew positif atau kemiringan ke kanan, hal ini dikarenakan median terletak lebih dekat pada kuartil pertama.

> Perbedaan boxplot dan histogram

Boxplot merupakan gambaran grafis yang menampilkan distribusi data secara ringkas dan komponennya berupa nilai minimal dan maksimal data, lalu median, kuartil dan outlier nya, sedangkan histogram merupakan grafik yang distribusinya menampilkan frekuensi dengan jumlah data secara interval dan ditampilkan dalam bentuk rangkaian batang vertikal. 




```python
#membuat scatterplot curah hujan dengan penjualan a

IQR = dh.quantile(0.75) - dh.quantile(0.25)
lowerboundH = dh.quantile(0.25) - 1.5 * IQR
upperboundH = dh.quantile(0.75) + 1.5 * IQR
outliers = (dh < lowerboundH) | (dh > upperboundH)
outliersh = dh[(dh<lowerboundH) | (dh>upperboundH)]
outliersh_data = data[dh.isin(outliersh)]
plt.figure(figsize=(8, 6))
plt.scatter(da, dh, label='Data Normal', color='blue')
plt.scatter(da[outliers], dh[outliers], label='Outlier', color='red')
plt.xlabel('Penjualan A (pcs)')
plt.ylabel('Curah Hujan (mm)')
plt.title('Scatterplot Curah Hujan vs Penjualan A dengan Outlier')
plt.legend()
print("Outliers : ")
print(outliersh_data)
plt.show()

#membuat scatterplot curah hujan dengan penjualan b

plt.figure(figsize=(8, 6))
plt.scatter(db, dh, label='Data Normal', color='blue')
plt.scatter(db[outliers], dh[outliers], label='Outlier', color='red')
plt.xlabel('Penjualan B (pcs)')
plt.ylabel('Curah Hujan (mm)')
plt.title('Scatterplot Curah Hujan vs Penjualan B dengan Outlier')
plt.legend()
plt.show()
```

Pemetaan menggunakan scatterplot mengenai korelasi curah hujan terhadap penjualan a dan b, menunjukkan beberapa outlier yang diidentifikasi dengan warna merah

> Bentuk Distribusi
- Ketidaksimetrian data (Skewness)
Ketiga data memiliki nilai ketidaksimetrian lebih dari nol yang menunjukkan penyimpangan simetri positif atau ke kanan. Ketiga data tersebut juga memiliki nilai rata-rata yang lebih tinggi dari mediannya. 
- Kurtosis
Data curah hujan memiliki nilai kurtosis yang sangat tinggi menandakan bahwa data tersebut sangat leptokurtik dan mempunyai nilai ekstrim lebih banyak dibandingkan distribusi normal.
Sedangkan data penjualan A dan B memiliki nilai kurtosis negatif yang menunjukkan bahwa kedua data itu platikurtik, menunjukkan bahwa data tersebut lebih menyebar dan memiliki nilai ekstrim yang lebih sedikit dibandingkan distribusi normal.

A. Standar Deviasi

B. Variansi dan rentang data
Menunjukkan bahwa data tersebar jauh dari rata-ratanya serta memiliki persebaran data dan variabilitas tinggi. Begitu pula dengan penjualan a dan b yang memiliki nilai variansi jauh lebih besar.
